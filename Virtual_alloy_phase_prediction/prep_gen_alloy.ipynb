```python
import pandas as pd
import os
import numpy as np 
```

```python
pwd = os.getcwd()
```

```python
df = pd.read_csv(pwd+"/gen_alloy_feats.csv")
df.head()
```

```python
# LOG1P TRANSFORMING REQUIRED COLUMNS
cols_to_transform = ['avg_dev_AtomicWeight', 
                     'mode_MeltingT', 
                     'avg_dev_Column',
                     'range_CovalentRadius', 
                     'avg_dev_CovalentRadius', 
                     'mode_Electronegativity',
                     'maximum_NpValence', 
                     'avg_dev_NpValence',
                     'avg_dev_NdValence',
                     'avg_dev_NValence', 
                     'mode_NValence', 
                     'avg_dev_NdUnfilled',
                     'minimum_GSvolume_pa',
                     'range_GSvolume_pa', 
                     'mode_GSvolume_pa']
df_tr = df.copy()
for c in cols_to_transform:
    df[c] = df[c].apply(np.log1p)
df.head()
```

```python
df.to_csv(pwd+"/gen_alloy_pp.csv", index=False)
```
