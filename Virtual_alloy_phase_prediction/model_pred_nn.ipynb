```python
import numpy as np
import pandas as pd
pd.set_option("display.max_columns", None)
from keras.models import load_model
from keras.utils import to_categorical
import os
import pickle
```

```python
pwd = os.getcwd()
```

```python
# LOAD DATAFRAME
df = pd.read_csv(pwd+"/gen_alloy_pp.csv")
print("Shape of dataframe:", df.shape)
df.head()
```

```python
# SELECT FEATURES
X = df.drop(["formula", "Probability"], axis=1, inplace=False)
print("Shape of feature dataframe:", X.shape)
X.head()
```

```python
# IMPORT SCALER AND LABEL ENCODER
scaler = pickle.load(open(pwd+"/scaler.pkl", "rb"))
lb_enc = pickle.load(open(pwd+"/label_enc.pkl", "rb"))
```

```python
# SCALE THE FEATURES
X_scaled = pd.DataFrame(scaler.transform(X), columns=X.columns)
X_scaled.head()
```

```python
# LOAD MODEL
model = load_model(pwd+"/nn_model.h5")
```

```python
# PREDICT ONE-HOT-ENCODED PHASE
phase_pred_logit = model.predict(X_scaled)
phase_pred_logit
```

```python
# CONVERT TO LABEL ENCODED PREDICTED PHASE
phase_pred_lb_enc = np.argmax(phase_pred_logit, axis=1)
phase_pred_lb_enc
```

```python
# DECODE USING LABEL_ENCODER
phase_pred = lb_enc.inverse_transform(phase_pred_lb_enc)
phase_pred
```

```python
df["predicted_phase"] = phase_pred
df.head(20)
```
